```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Assuming this data is given to find out in whatways we can improve sales n business in all aspects, the below observations are made.

```python
data = pd.read_csv("company.csv")
```

```python
data.size
```

```python
data.head()
```

```python
data.groupby(['Company','Place']).count().fillna(0).sort_values(by="Place",ascending=True)
```

```python
len(data.Company.unique())
```

```python
len(data.Place.unique())
```

```python
data.groupby(['Company','Place']).count().sort_values("Place",ascending=True)
```

```python
sal = data.groupby(['Company','Place']).mean().sort_values("Place",ascending=True).drop('Age',axis=1).drop('Gender',axis=1)
```

```python
sal.sort_values(by='Salary',ascending=False)
```

```python
data.groupby(['Place','Gender']).mean()
```

```python
plt.figure(figsize = (16,6))
plt.title("Company Data")
sns.heatmap(data.corr(),annot = True,vmin=-1,vmax=1,center=0)
plt.xlabel("Company")
```

```python
sns.pairplot(data=data)
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x= data['Age'],y= data['Salary'])
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x= data['Age'],y= data['Salary'])
```

#### The  general trend in relation with age and salary is quite comparable. As the age (as in experience) increases the  salar also tends to increase.

```python
plt.figure(figsize=(16,6))
sns.boxplot(x= data['Gender'],y= data['Salary'])
```

### From this graph we can conclude that avg salary for women is less compared to men

```python
plt.figure(figsize=(16,6))
sns.distplot(a=data['Salary'],bins=100)
```

```python
plt.figure(figsize=(16,6))
sns.swarmplot(x = data['Age'], y =data['Salary'], hue= data['Gender'] )
```

```python
plt.figure(figsize=(16,6))
sns.boxplot(data = data,orient = 'h',color = '.8')
```
